```python
import plotly
import plotly.graph_objs as go
import re
plotly.offline.init_notebook_mode(connected = True)
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))

def add_file_to_loss(name, loss):
    re_loss = re.compile("(?<=loss )([^\s]+)")
    with open(name, 'r') as f:
        loss[-1]['train_loss'] = []
        loss[-1]['test_loss'] = []
        for line in f:
            try:
                if line.startswith('Train'):
                    loss[-1]['train_loss'].append(float(re_loss.search(line).group(0)))
                elif line.startswith('Test'):
                    loss[-1]['test_loss'].append(float(re_loss.search(line).group(0)))
            except AttributeError: pass
```

```python
import glob
import os
loss=[]
for name in set([pp for p in ('log/naive*',
                             'log/n*') for pp in glob.glob(p)]):
    if any(n in name for n in ['fail', 'Nan', '20']): continue
    loss.append({}); loss[-1]['name'] = os.path.basename(name)[:-4]; add_file_to_loss(name, loss); print(name)
```

```python
data = [go.Scatter(name= l['name'], x = list(range(len(l['train_loss']))), y = l['train_loss']) for l in loss]
#data += [go.Scatter(name= l['name'], x = range(len(l['train_loss'])), y = l['train_loss']) for l in loss]
fig = go.Figure(data=data, layout=go.Layout(legend=dict(orientation='v')))

plotly.offline.iplot(fig)
```
