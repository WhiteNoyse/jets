```python
Z = numpy.zeros((X.shape[0], X.shape[0]))
for i in range(X.shape[0]):
    for j in range(i, X.shape[0]):
        z_ = numpy.sqrt(((Y[i].mean(1) - Y[j].mean(1))**2).sum())
        if z_ > 0.:
            Z[i,j] =1./z_
#         Z[i,j] =numpy.sqrt(((Y[i].mean(1) - Y[j].mean(1))**2).sum())
```
